```python
import requests
import pandas as pd

# NHL API for skater stats leaders
nhl_url = "https://api-web.nhle.com/v1/skater-stats-leaders/current?categories=points&limit=25"

# Fetch the data
response = requests.get(nhl_url)
data = response.json()

# Extract and format top 25
top_25 = data.get("points", [])[:25]
top_players = [{
    "id": p["id"],
    "name": f"{p['firstName']['default']} {p['lastName']['default']}",
    "team": p["teamAbbrev"],
    "position": p["position"],
    "points": p["value"],
    "headshot": p["headshot"]
} for p in top_25]

df = pd.DataFrame(top_players)
print(df)
```

```python
import asyncio
from playwright.async_api import async_playwright

async def scrape_fight_count(player_slug):
    url = f"https://www.hockeyfights.com/players/{player_slug}"

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto(url)

        await page.wait_for_selector("text=Fight Card", timeout=10000)

        content = await page.content()
        print(content[:3000])  # show some initial HTML so we can parse it later

        await browser.close()

# Run the async function properly in Jupyter
await scrape_fight_count("tom-wilson")
```
